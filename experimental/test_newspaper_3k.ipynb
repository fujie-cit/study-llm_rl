```python
import newspaper
```

```python
url = "https://www.bloomberg.com/jp/technology"
```

```python
website = newspaper.build(url, memoize_articles=False, MAX_SUMMARY=300)
```

```python
for article in website.articles:
    try:
        article.download()
        article.parse()
        print(article.title)
        print(article.text)
        print("-----")
    except Exception as e:
        print(f"Error processing article: {e}")
```
