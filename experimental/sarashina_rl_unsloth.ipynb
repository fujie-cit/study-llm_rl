```python
# -----------------------------
# 2) PEFT (LoRA) ã‚’ policy å´ã ã‘ã«é©ç”¨
#   - ref_model ã¯å›ºå®šã®ã¾ã¾ï¼ˆLoRA ã™ã‚‰è¼‰ã›ãªã„ã®ãŒã€Œæœ€å°ã€ï¼‰
# -----------------------------
lora_config = LoraConfig(
    r=8,                     # 8 or 16 ãŒã¾ãšã¯ç„¡é›£
    lora_alpha=16,           # é€šå¸¸ r ã® 2 å€
    lora_dropout=0.05,       # DPO ã§ã¯ 0.0ã€œ0.1
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
    ],
)

model = get_peft_model(base_model, lora_config)
```

```python
# -----------------------------
# 3) DPO è¨­å®šï¼ˆÎ²ã€ãƒãƒƒãƒç­‰ï¼‰
#   - beta ãŒ 2.5 ã® Î²
# -----------------------------
dpo_args = DPOConfig(
    output_dir="./dpo_lora_out",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=1,
    learning_rate=5e-5,
    num_train_epochs=1,
    beta=0.1,
    logging_steps=1,
    save_steps=50,
    fp16=torch.cuda.is_available(),
    # å®‰å®šæ€§ã®ãŸã‚ã« max_length / max_prompt_length ã‚’å…¥ã‚Œã‚‹ã®ãŒä¸€èˆ¬çš„
    max_length=512,
    max_prompt_length=256,
)
```
